```python
import pandas as pd
import numpy as np
import seaborn
from sklearn.preprocessing import StandardScaler
```

```python
data=pd.read_csv('/kaggle/input/obfuscated-malware-memory-2022-cic/Obfuscated-MalMem2022.csv')
```

```python
data.isnull().sum()
```

```python
data.Class.value_counts()
```

```python
data=data.drop('Category',axis=1)

data
```

```python
data=data.drop('pslist.nprocs64bit',axis=1)
data=data.drop('handles.nport',axis=1)
data=data.drop('handles.avg_handles_per_proc',axis=1)
data=data.drop('ldrmodules.not_in_load_avg',axis=1)
data=data.drop('ldrmodules.not_in_mem_avg',axis=1)
data=data.drop('ldrmodules.not_in_init_avg',axis=1)
data=data.drop('malfind.commitCharge',axis=1)
data=data.drop('malfind.protection',axis=1)
data=data.drop('malfind.uniqueInjections',axis=1)
data
```

```python
data=data.drop('svcscan.fs_drivers',axis=1)
data=data.drop('callbacks.ngeneric',axis=1)
data
```

```python
x_col=data.columns.to_list()
x_col.pop(-1)
x_data=data[x_col]
y_col='Class'
```

```python
from sklearn.model_selection import train_test_split
train_x,test_x,train_y,test_y = train_test_split(data[x_col],data[y_col].values,test_size=0.3)
train_x.shape
```

```python
# Normalize the data
sc = StandardScaler()
normed_train_data = pd.DataFrame(sc.fit_transform(train_x), columns = x_col)
normed_test_data = pd.DataFrame(sc.fit_transform(test_x), columns = x_col)
```

```python
from sklearn.ensemble import RandomForestClassifier
rf=RandomForestClassifier(criterion='entropy')
rf.fit(train_x,train_y)

y_pred= rf.predict(test_x)
```

```python
from sklearn.metrics import accuracy_score
# Calculate the accuracy of the model
accuracy = accuracy_score(test_y, y_pred)
print("Accuracy:", accuracy)
from sklearn.metrics import precision_score, recall_score, f1_score

# Assuming you have predicted labels (y_pred) and true labels (y_test)

# Calculate precision
precision = precision_score(test_y, y_pred, pos_label='Malware')

# Calculate recall
recall = recall_score(test_y, y_pred, pos_label='Malware')

# Calculate F1 score
f1 = f1_score(test_y, y_pred, pos_label='Malware')

# Print the evaluation metrics
print("Precision:", precision)
print("Recall:", recall)
print("F1 Score:", f1)
```
